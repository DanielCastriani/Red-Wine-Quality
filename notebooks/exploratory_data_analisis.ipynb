```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
from utils.ecdf import ecdf_plot

from dataset.make_dataset import make_dataset
```

```python
red_wine = make_dataset()
```

```python
fig = plt.figure(figsize=(35, 15))
gs = GridSpec(3, 4)


sns.histplot(red_wine['fixed_acidity'], kde=True, ax=fig.add_subplot(gs[0, 0], title='fixed_acidity'))
sns.histplot(red_wine['volatile_acidity'], kde=True, ax=fig.add_subplot(gs[0, 1], title='volatile_acidity'))
sns.histplot(red_wine['citric_acid'], kde=True, ax=fig.add_subplot(gs[0, 2], title='citric_acid'))
sns.histplot(red_wine['residual_sugar'], kde=True, ax=fig.add_subplot(gs[0, 3], title='residual_sugar'))

sns.histplot(red_wine['chlorides'], kde=True, ax=fig.add_subplot(gs[1, 0], title='chlorides'))
sns.histplot(red_wine['free_sulfur_dioxide'], kde=True, ax=fig.add_subplot(gs[1, 1], title='free_sulfur_dioxide'))
sns.histplot(red_wine['total_sulfur_dioxide'], kde=True, ax=fig.add_subplot(gs[1, 2], title='total_sulfur_dioxide'))
sns.histplot(red_wine['density'], kde=True, ax=fig.add_subplot(gs[1, 3], title='density'))

sns.histplot(red_wine['pH'], kde=True, ax=fig.add_subplot(gs[2, 0], title='pH'))
sns.histplot(red_wine['sulphates'], kde=True, ax=fig.add_subplot(gs[2, 1], title='sulphates'))
sns.histplot(red_wine['alcohol'], kde=True, ax=fig.add_subplot(gs[2, 2], title='alcohol'))
sns.histplot(red_wine['quality'], kde=True, ax=fig.add_subplot(gs[2, 3], title='quality'))

plt.show()
```

```python
fig = plt.figure(figsize=(35, 15))
gs = GridSpec(3, 4)


ecdf_plot(red_wine['fixed_acidity'], ax=fig.add_subplot(gs[0, 0], title='fixed_acidity'), show=False)
ecdf_plot(red_wine['volatile_acidity'], ax=fig.add_subplot(gs[0, 1], title='volatile_acidity'), show=False)
ecdf_plot(red_wine['citric_acid'], ax=fig.add_subplot(gs[0, 2], title='citric_acid'), show=False)
ecdf_plot(red_wine['residual_sugar'], ax=fig.add_subplot(gs[0, 3], title='residual_sugar'), show=False)

ecdf_plot(red_wine['chlorides'], ax=fig.add_subplot(gs[1, 0], title='chlorides'), show=False)
ecdf_plot(red_wine['free_sulfur_dioxide'], ax=fig.add_subplot(gs[1, 1], title='free_sulfur_dioxide'), show=False)
ecdf_plot(red_wine['total_sulfur_dioxide'], ax=fig.add_subplot(gs[1, 2], title='total_sulfur_dioxide'), show=False)
ecdf_plot(red_wine['density'], ax=fig.add_subplot(gs[1, 3], title='density'), show=False)

ecdf_plot(red_wine['pH'], ax=fig.add_subplot(gs[2, 0], title='pH'), show=False)
ecdf_plot(red_wine['sulphates'], ax=fig.add_subplot(gs[2, 1], title='sulphates'), show=False)
ecdf_plot(red_wine['alcohol'], ax=fig.add_subplot(gs[2, 2], title='alcohol'), show=False)
ecdf_plot(red_wine['quality'], ax=fig.add_subplot(gs[2, 3], title='quality'), show=False)

plt.show()
```
